```python
df_10_group_1 = df_10.groupby(['행정동'])['거주인구'].sum().reset_index()
```

```python
df_10_group_1.loc[df_10_group_1['행정동'] == '가수원동']['거주인구'][0]
```

```python
def ratio_resid(i):
    data = df_10_group.loc[df_10_group['행정동'] == i].pivot(index= '행정동',columns= '연령대', values = '거주인구')
    value = data['10대'][0] * 5 + data['10대 미만'][0] * 15 + data['20대'][0] * 25 + data['30대'][0] * 35 + data['40대'][0] * 45 + data['50대'][0] * 55 + data['60대'][0] * 65 + data['70대 이상'][0] * 75
    resid = list(df_10_group_1.loc[df_10_group_1['행정동'] == i]['거주인구'])[0]
    ratio = value / resid
    return [i, int(ratio)]
```

```python
list_dong = []
for i in df_10_group['행정동'].unique():
    b = ratio_resid(i)
    list_dong.append(b)
```

```python
a = pd.DataFrame(list_dong, columns = ['행정동','평균 나이'])
```
